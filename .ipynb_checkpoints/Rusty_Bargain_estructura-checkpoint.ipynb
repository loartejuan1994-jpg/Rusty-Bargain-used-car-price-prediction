# Rusty Bargain: Predicción del valor de mercado de autos usados

## Descripción del proyecto

El servicio de venta de autos usados **Rusty Bargain** está desarrollando una aplicación para atraer nuevos clientes. Gracias a esa app, los usuarios pueden averiguar rápidamente el valor de mercado de su coche. Se cuenta con acceso al historial de datos: especificaciones técnicas, versiones de equipamiento y precios de vehículos.

El objetivo es construir un modelo que determine el valor de mercado de un auto usado a partir de sus características.

## Objetivos

- Preparar y limpiar los datos disponibles para el modelado.
- Entrenar distintos modelos de regresión (lineal, basados en árboles y de potenciación de gradiente) para predecir el precio de los autos.
- Evaluar y comparar los modelos considerando tres criterios clave para la empresa:
  1. **Calidad de la predicción** (medida con la raíz del error cuadrático medio, RECM/RMSE).
  2. **Velocidad de la predicción** (tiempo de inferencia).
  3. **Tiempo requerido para el entrenamiento**.
- Seleccionar el modelo que ofrezca el mejor equilibrio entre estas tres dimensiones.

## Descripción de los datos

El dataset contiene la siguiente información (columnas típicas del dataset `car_data.csv`):

- `DateCrawled` — fecha en la que se descargó el perfil de la base de datos
- `VehicleType` — tipo de carrocería del vehículo
- `RegistrationYear` — año de matriculación del vehículo
- `Gearbox` — tipo de caja de cambios
- `Power` — potencia (CV)
- `Model` — modelo del vehículo
- `Mileage` — kilometraje (medido en km)
- `RegistrationMonth` — mes de matriculación del vehículo
- `FuelType` — tipo de combustible
- `Brand` — marca del vehículo
- `NotRepaired` — vehículo con o sin reparación
- `DateCreated` — fecha de creación del perfil
- `NumberOfPictures` — número de fotos del vehículo
- `PostalCode` — código postal del propietario
- `LastSeen` — fecha de la última actividad del usuario

**Objetivo (target):**
- `Price` — precio (en euros)

## Plan de trabajo

1. **Preparación de datos**: carga, exploración, limpieza de valores ausentes/atípicos, codificación de variables categóricas y división en conjuntos de entrenamiento/validación/prueba.
2. **Entrenamiento del modelo**: entrenamiento de un modelo base (regresión lineal) y de modelos más complejos (árbol de decisión, bosque aleatorio, LightGBM y/o CatBoost), midiendo tiempo de entrenamiento y de predicción.
3. **Análisis del modelo**: comparación de RECM, tiempo de entrenamiento y tiempo de predicción entre todos los modelos, y selección del mejor modelo.
4. **Conclusiones**: resumen de hallazgos y recomendación final para Rusty Bargain.


## Preparación de datos

```python
# Librerías generales
import numpy as np
import pandas as pd
import time
import warnings

warnings.filterwarnings('ignore')

# Visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocesamiento y validación
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.metrics import mean_squared_error

# Modelos
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Modelos de potenciación de gradiente (instalar si es necesario)
# !pip install lightgbm catboost --quiet
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

RANDOM_STATE = 12345
```

```python
# Carga de datos
df = pd.read_csv('/datasets/car_data.csv')
df.head()
```

```python
# Información general del dataset
df.info()
```

```python
# Estadísticas descriptivas
df.describe(include='all').T
```

### Revisión y limpieza de datos

En esta sección se identifican y tratan:
- Valores ausentes en columnas categóricas (`VehicleType`, `Gearbox`, `Model`, `FuelType`, `NotRepaired`).
- Valores atípicos o poco realistas (por ejemplo, `Price` igual a 0, `RegistrationYear` fuera de un rango razonable, `Power` igual a 0 o extremadamente alto).
- Columnas irrelevantes para la predicción del precio (fechas de rastreo, número de fotos, código postal, etc.).


```python
# Revisión de valores ausentes
df.isna().sum()
```

```python
# Eliminación de columnas irrelevantes para el modelo
columns_to_drop = ['DateCrawled', 'DateCreated', 'NumberOfPictures', 'PostalCode', 'LastSeen']
df = df.drop(columns=[c for c in columns_to_drop if c in df.columns])
df.head()
```

```python
# Tratamiento de valores ausentes en variables categóricas
categorical_cols = ['VehicleType', 'Gearbox', 'Model', 'FuelType', 'NotRepaired']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna('unknown')

df.isna().sum()
```

```python
# Filtrado de valores atípicos / poco realistas
df = df[(df['Price'] > 100) & (df['Price'] < 20000)]
df = df[(df['RegistrationYear'] >= 1990) & (df['RegistrationYear'] <= 2019)]
df = df[(df['Power'] > 0) & (df['Power'] < 500)]

df = df.reset_index(drop=True)
df.shape
```

```python
# Codificación de variables categóricas (Ordinal Encoding, adecuado para modelos basados en árboles)
cat_features = df.select_dtypes(include='object').columns.tolist()

encoder = OrdinalEncoder()
df[cat_features] = encoder.fit_transform(df[cat_features])

df.head()
```

```python
# Separación de características y objetivo
features = df.drop(columns=['Price'])
target = df['Price']

# División en entrenamiento (60%), validación (20%) y prueba (20%)
features_train, features_temp, target_train, target_temp = train_test_split(
    features, target, test_size=0.4, random_state=RANDOM_STATE
)
features_valid, features_test, target_valid, target_test = train_test_split(
    features_temp, target_temp, test_size=0.5, random_state=RANDOM_STATE
)

print(features_train.shape, features_valid.shape, features_test.shape)
```

```python
# Escalado de características numéricas (necesario para la regresión lineal)
numeric_features = ['RegistrationYear', 'Power', 'Mileage', 'RegistrationMonth']
numeric_features = [c for c in numeric_features if c in features.columns]

scaler = StandardScaler()
features_train_scaled = features_train.copy()
features_valid_scaled = features_valid.copy()
features_test_scaled = features_test.copy()

features_train_scaled[numeric_features] = scaler.fit_transform(features_train[numeric_features])
features_valid_scaled[numeric_features] = scaler.transform(features_valid[numeric_features])
features_test_scaled[numeric_features] = scaler.transform(features_test[numeric_features])
```

## Entrenamiento del modelo

Se entrenan distintos modelos, midiendo en cada caso el **tiempo de entrenamiento**, el **tiempo de predicción** y el **RECM (RMSE)** sobre el conjunto de validación. Los resultados se almacenan en un diccionario para compararlos posteriormente.


```python
results = {}

def evaluar_modelo(nombre, modelo, X_train, y_train, X_valid, y_valid):
    start_train = time.time()
    modelo.fit(X_train, y_train)
    train_time = time.time() - start_train

    start_pred = time.time()
    predictions = modelo.predict(X_valid)
    predict_time = time.time() - start_pred

    rmse = mean_squared_error(y_valid, predictions, squared=False)

    results[nombre] = {
        'RMSE': rmse,
        'Tiempo de entrenamiento (s)': train_time,
        'Tiempo de predicción (s)': predict_time
    }

    print(f'{nombre}: RMSE={rmse:.2f} | Entrenamiento={train_time:.2f}s | Predicción={predict_time:.4f}s')
    return modelo
```

### Modelo base: Regresión Lineal

```python
modelo_lr = LinearRegression()
evaluar_modelo(
    'Regresión Lineal',
    modelo_lr,
    features_train_scaled, target_train,
    features_valid_scaled, target_valid
)
```

### Árbol de Decisión

```python
modelo_tree = DecisionTreeRegressor(random_state=RANDOM_STATE, max_depth=10)
evaluar_modelo(
    'Árbol de Decisión',
    modelo_tree,
    features_train, target_train,
    features_valid, target_valid
)
```

### Bosque Aleatorio

```python
modelo_rf = RandomForestRegressor(
    random_state=RANDOM_STATE,
    n_estimators=100,
    max_depth=12,
    n_jobs=-1
)
evaluar_modelo(
    'Bosque Aleatorio',
    modelo_rf,
    features_train, target_train,
    features_valid, target_valid
)
```

### LightGBM

```python
modelo_lgbm = LGBMRegressor(
    random_state=RANDOM_STATE,
    n_estimators=500,
    learning_rate=0.05
)
evaluar_modelo(
    'LightGBM',
    modelo_lgbm,
    features_train, target_train,
    features_valid, target_valid
)
```

### CatBoost

```python
modelo_cb = CatBoostRegressor(
    random_state=RANDOM_STATE,
    iterations=500,
    learning_rate=0.05,
    depth=8,
    verbose=False
)
evaluar_modelo(
    'CatBoost',
    modelo_cb,
    features_train, target_train,
    features_valid, target_valid
)
```

## Análisis del modelo

Se comparan todos los modelos entrenados según los tres criterios de interés para Rusty Bargain: **calidad (RECM)**, **velocidad de predicción** y **tiempo de entrenamiento**.


```python
resultados_df = pd.DataFrame(results).T
resultados_df = resultados_df.sort_values('RMSE')
resultados_df
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

resultados_df['RMSE'].plot(kind='barh', ax=axes[0], color='steelblue')
axes[0].set_title('RECM (RMSE) por modelo')
axes[0].set_xlabel('RMSE')

resultados_df['Tiempo de entrenamiento (s)'].plot(kind='barh', ax=axes[1], color='darkorange')
axes[1].set_title('Tiempo de entrenamiento por modelo')
axes[1].set_xlabel('Segundos')

resultados_df['Tiempo de predicción (s)'].plot(kind='barh', ax=axes[2], color='seagreen')
axes[2].set_title('Tiempo de predicción por modelo')
axes[2].set_xlabel('Segundos')

plt.tight_layout()
plt.show()
```

### Selección del mejor modelo y prueba final

Una vez identificado el modelo con el mejor equilibrio entre calidad y velocidad (típicamente LightGBM o CatBoost), se evalúa su desempeño final sobre el **conjunto de prueba**, que no fue utilizado en ninguna etapa anterior.


```python
# Ejemplo: evaluación final del modelo seleccionado sobre el conjunto de prueba
mejor_modelo = modelo_lgbm  # sustituir por el modelo con mejor desempeño

start_pred = time.time()
predicciones_test = mejor_modelo.predict(features_test)
tiempo_prediccion_test = time.time() - start_pred

rmse_test = mean_squared_error(target_test, predicciones_test, squared=False)

print(f'RMSE en el conjunto de prueba: {rmse_test:.2f}')
print(f'Tiempo de predicción en prueba: {tiempo_prediccion_test:.4f}s')
```

```python
# Prueba de cordura: comparación contra un modelo dummy (predicción de la media)
from sklearn.dummy import DummyRegressor

modelo_dummy = DummyRegressor(strategy='mean')
evaluar_modelo(
    'Modelo Dummy (referencia)',
    modelo_dummy,
    features_train, target_train,
    features_valid, target_valid
)
```

## Conclusiones

*(Completar con base en los resultados numéricos obtenidos al ejecutar el notebook con los datos reales)*

- **Calidad de la predicción**: los modelos de potenciación de gradiente (LightGBM, CatBoost) suelen superar en RECM a la regresión lineal y al árbol de decisión individual, gracias a su capacidad de capturar relaciones no lineales entre las características y el precio.
- **Velocidad de entrenamiento**: la regresión lineal es la más rápida de entrenar, seguida del árbol de decisión; los modelos de ensamble (bosque aleatorio, LightGBM, CatBoost) requieren más tiempo de entrenamiento, siendo generalmente CatBoost el más costoso y LightGBM uno de los más eficientes dentro de los modelos de boosting.
- **Velocidad de predicción**: todos los modelos entrenados ofrecen tiempos de predicción muy bajos, adecuados para su uso en una aplicación en tiempo real; las diferencias entre ellos suelen ser marginales frente a las diferencias en tiempo de entrenamiento.
- **Modelo recomendado**: se recomienda utilizar **[nombre del modelo con mejor RECM y tiempos razonables]**, ya que ofrece el mejor equilibrio entre calidad de predicción, tiempo de entrenamiento y velocidad de predicción, cumpliendo con los tres criterios solicitados por Rusty Bargain.
- **Próximos pasos**: explorar ingeniería de características adicional (por ejemplo, antigüedad del vehículo a partir de `RegistrationYear`), validación cruzada más exhaustiva y ajuste fino de hiperparámetros para mejorar aún más el RECM sin sacrificar demasiado la velocidad de entrenamiento.
